```python
def CalculPrix(Louer_Acheter, M2, Lieu, EtatGeneral, Meuble, Disposition, Orientation, Hauteur, Locaux, Surplus):
    if Louer_Acheter == "Louer":
        #Lieu=Arrondissement
        if Lieu == 1:
            MCarre= 33
        elif Lieu == 2:
            MCarre= 30
        elif Lieu == 3:
            MCarre= 32
        elif Lieu == 4:
            MCarre= 33
        elif Lieu == 5:
            MCarre= 32
        elif Lieu == 6:
            MCarre= 33
        elif Lieu == 7:
            MCarre= 32
        elif Lieu == 8:
            MCarre= 31
        elif Lieu == 9:
            MCarre= 31
        elif Lieu == 10:
            MCarre= 30
        elif Lieu == 11:
            MCarre= 30
        elif Lieu == 12:
            MCarre= 31
        elif Lieu == 13:
            MCarre= 30
        elif Lieu == 14:
            MCarre= 30
        elif Lieu == 15:
            MCarre= 29
        elif Lieu == 16:
            MCarre= 30
        elif Lieu == 17:
            MCarre= 30
        elif Lieu == 18:
            MCarre= 28
        elif Lieu == 19:
            MCarre= 28
        elif Lieu == 20:
            MCarre= 29
        elif Lieu =="Rien":
            return "Vous devais remplir ici."
        else:
            return "Error"
        #print("PrixLouer",MCarre)
        
    elif Louer_Acheter == "Acheter":
        #Lieu=Arrondissement
        if Lieu == 1:
            MCarre= 10200
        elif Lieu == 2:
            MCarre= 9600
        elif Lieu == 3:
            MCarre= 9800
        elif Lieu == 4:
            MCarre= 10200
        elif Lieu == 5:
            MCarre= 11000
        elif Lieu == 6:
            MCarre= 11000
        elif Lieu == 7:
            MCarre= 10600
        elif Lieu == 8:
            MCarre= 9700
        elif Lieu == 9:
            MCarre= 10200
        elif Lieu == 10:
            MCarre= 10200
        elif Lieu == 11:
            MCarre= 8900
        elif Lieu == 12:
            MCarre= 8600
        elif Lieu == 13:
            MCarre= 7400
        elif Lieu == 14:
            MCarre= 8700
        elif Lieu == 15:
            MCarre= 7300
        elif Lieu == 16:
            MCarre= 9000
        elif Lieu == 17:
            MCarre= 8300
        elif Lieu == 18:
            MCarre= 8700
        elif Lieu == 19:
            MCarre= 8100
        elif Lieu == 20:
            MCarre= 7900
        elif Lieu =="Rien":
            return "Vous devais remplir ici."
        else: 
            return "Error"
        #print("PrixAcheter",MCarre)
        
    if EtatGeneral == "Propre":
        EtatGeneral = 0
    elif EtatGeneral == "Sale":
        EtatGeneral =0.05
    elif EtatGeneral=="Rien":
        EtatGeneral=0
    else: 
        return "Error"
    #print("EtatGeneral",EtatGeneral,-MCarre*EtatGeneral)
    
    if Meuble == "Oui":
        Meuble=0.05
    elif Meuble == "Non":
        Meuble=0
    elif Meuble =="Rien":
        Meuble=0
    else: 
        return "Error"   
    #print("Meuble",Meuble,MCarre*Meuble)
    
    if Disposition == "Ville":
        Disposition =0.1
    elif Disposition == "Isole":
        Disposition =-0.05
    elif Disposition =="Rien":
        Disposition =0
    else: 
        return "Error"  
    #print("Disposition",Disposition,MCarre*Disposition)

    Orientation=0
    if Orientation == "Est":
        Orientation=0.05
    #print("Orientation",Orientation,MCarre*Orientation)
    
    if Locaux == "Escalier" and Hauteur>=3:
        Hauteur=0.15
    elif Locaux == "Escalier":
        Hauteur=0.05
    elif Locaux == "Ascenseur" and Hauteur>=3:
        Hauteur=0.05
    elif Locaux == "Ascenseur":
        Hauteur=-0.05
    elif Locaux == "Rien":
        Hauteur==0
    else:
        return "Error"  
    #print("Hauteur",Hauteur,-MCarre*Hauteur)
    
    CptSurplus=0
    if Surplus!="Rien":
        for i in range (0,Surplus):
            CptSurplus+=1
        Surplus=CptSurplus*0.05
    elif Surplus=="Rien":
        Surplus=0
    #print("Surplus",Surplus,MCarre*Surplus)
        
    Prix=0  
    #print("MCarre",MCarre,"M2",M2,"PrixMCarre",MCarre*M2)
    MCarre=MCarre*M2
    Prix=MCarre-MCarre*EtatGeneral+MCarre*Disposition+MCarre*Meuble+MCarre*Orientation-MCarre*Hauteur+MCarre*Surplus
    return "Le prix du batiment est:",Prix
```

```python
def VautMieux(Rentabilite, M2, Lieu, EtatGeneral, Meuble, Disposition, Orientation, Hauteur, Locaux, Surplus):
    LouerIni=CalculPrix("Louer", M2, Lieu, EtatGeneral, Meuble, Disposition, Orientation, Hauteur, Locaux, Surplus)
    Acheter=CalculPrix("Acheter", M2, Lieu, EtatGeneral, Meuble, Disposition, Orientation, Hauteur, Locaux, Surplus)
    RentabiliteMois=Rentabilite*12
    Louer=LouerIni[1]*RentabiliteMois
    Acheter=Acheter[1]
    print("Si l'on loue:",Louer,"sur",Rentabilite,"ans.")
    print("Si l'on achete",Acheter,"sur",Rentabilite,"ans.")
    if Louer < Acheter:
        while Louer < Acheter:
            Louer=Louer+LouerIni[1]
            RentabiliteMois+=1
        Rentabilite=int(RentabiliteMois/12)
        return "Il vaut mieux louer dans ce cas là ou acheter et relouer pendant",Rentabilite,"ans"
    elif Louer > Acheter:
        return "Il vaut mieux acheter dans ce cas là"
    else:
        return "C'est identique donc sois attendre un peu plus longtemps et acheter sois louer"
```
