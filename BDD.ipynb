```python
def CalculPrix(Louer_Acheter, M2, Lieu, EtatGeneral, Meuble, Disposition, Orientation, Hauteur, Locaux, Surplus, NbPiece, NbBain, NbToilette):
    if Louer_Acheter == "Louer":
        #Lieu=Arrondissement
        if Lieu == 1:
            MCarre= 33
        elif Lieu == 2:
            MCarre= 30
        elif Lieu == 3:
            MCarre= 32
        elif Lieu == 4:
            MCarre= 33
        elif Lieu == 5:
            MCarre= 32
        elif Lieu == 6:
            MCarre= 33
        elif Lieu == 7:
            MCarre= 32
        elif Lieu == 8:
            MCarre= 31
        elif Lieu == 9:
            MCarre= 31
        elif Lieu == 10:
            MCarre= 30
        elif Lieu == 11:
            MCarre= 30
        elif Lieu == 12:
            MCarre= 31
        elif Lieu == 13:
            MCarre= 30
        elif Lieu == 14:
            MCarre= 30
        elif Lieu == 15:
            MCarre= 29
        elif Lieu == 16:
            MCarre= 30
        elif Lieu == 17:
            MCarre= 30
        elif Lieu == 18:
            MCarre= 28
        elif Lieu == 19:
            MCarre= 28
        elif Lieu == 20:
            MCarre= 29
        elif Lieu =="Rien":
            return "Vous devais remplir ici."
        else:
            return "Error"
        #print("PrixLouer",MCarre)
        
    elif Louer_Acheter == "Acheter":
        #Lieu=Arrondissement
        if Lieu == 1:
            MCarre= 10200
        elif Lieu == 2:
            MCarre= 9600
        elif Lieu == 3:
            MCarre= 9800
        elif Lieu == 4:
            MCarre= 10200
        elif Lieu == 5:
            MCarre= 11000
        elif Lieu == 6:
            MCarre= 11000
        elif Lieu == 7:
            MCarre= 10600
        elif Lieu == 8:
            MCarre= 9700
        elif Lieu == 9:
            MCarre= 10200
        elif Lieu == 10:
            MCarre= 10200
        elif Lieu == 11:
            MCarre= 8900
        elif Lieu == 12:
            MCarre= 8600
        elif Lieu == 13:
            MCarre= 7400
        elif Lieu == 14:
            MCarre= 8700
        elif Lieu == 15:
            MCarre= 7300
        elif Lieu == 16:
            MCarre= 9000
        elif Lieu == 17:
            MCarre= 8300
        elif Lieu == 18:
            MCarre= 8700
        elif Lieu == 19:
            MCarre= 8100
        elif Lieu == 20:
            MCarre= 7900
        elif Lieu =="Rien":
            return "Vous devais remplir ici."
        else: 
            return "Error"
        #print("PrixAcheter",MCarre)
        
    if EtatGeneral == "Propre":
        EtatGeneral = 0
    elif EtatGeneral == "Sale":
        EtatGeneral =0.05
    elif EtatGeneral=="Rien":
        EtatGeneral=0
    else: 
        return "Error"
    #print("EtatGeneral",EtatGeneral,-MCarre*EtatGeneral)
    
    if Meuble == "Oui":
        Meuble=0.05
    elif Meuble == "Non":
        Meuble=0
    elif Meuble =="Rien":
        Meuble=0
    else: 
        return "Error"   
    #print("Meuble",Meuble,MCarre*Meuble)
    
    if Disposition == "Ville":
        Disposition =0.1
    elif Disposition == "Isole":
        Disposition =-0.05
    elif Disposition =="Rien":
        Disposition =0
    else: 
        return "Error"  
    #print("Disposition",Disposition,MCarre*Disposition)

    Orientation=0
    if Orientation == "Est":
        Orientation=0.05
    #print("Orientation",Orientation,MCarre*Orientation)
    
    if Locaux == "Escalier" and Hauteur>=3:
        Hauteur=0.15
    elif Locaux == "Escalier":
        Hauteur=0.05
    elif Locaux == "Ascenseur" and Hauteur>=3:
        Hauteur=0.05
    elif Locaux == "Ascenseur":
        Hauteur=-0.05
    elif Locaux == "Rien":
        Hauteur==0
    else:
        return "Error"  
    #print("Hauteur",Hauteur,-MCarre*Hauteur)
    
    CptSurplus=0
    if Surplus!="Rien":
        for i in range (0,Surplus):
            CptSurplus+=1
        Surplus=CptSurplus*0.05
    elif Surplus=="Rien":
        Surplus=0
    #print("Surplus",Surplus,MCarre*Surplus)
    
    PNbPiece=0
    if M2/NbPiece > 8:
        for i in range (2,NbPiece):
            PNbPiece+=0.05
    PNbBain=0
    if NbBain > 1:
        PNbBain+=0.05
        for i in range (0,NbBain):
            PNbBain+=0.05
            
    PNbToilette=0
    if NbToilette > 1:
        PNbToilette+=0.05
        for i in range (0,NbToilette):
            PNbToilette+=0.05
            
    Prix=0  
    #print("MCarre",MCarre,"M2",M2,"PrixMCarre",MCarre*M2)
    MCarre=MCarre*M2
    Prix=MCarre-MCarre*EtatGeneral+MCarre*Disposition+MCarre*Meuble+MCarre*Orientation-MCarre*Hauteur+MCarre*Surplus+MCarre*PNbPiece+MCarre*PNbBain+MCarre*PNbToilette
    return Prix
```

```python
#On charge les données

dataframe = pd.read_csv('BDDLASTML.csv', engine='python', )
dataset = dataframe.values

#On donne ce que l'on doit trouver (y) et avec quoi (X), prix antérieur dans le futur devient prix futur

X = dataframe.drop('Prix', axis=1)
y = dataframe['Prix']
#On définit la MAPE qui sert quantifier l'erreur: pourcentage moyen de différence

def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)   
    a = np.array(y_true)
    b = np.array(y_pred)
    mean = (a + b)/2.0
    return 100*np.mean(np.abs((y_true - y_pred) / np.maximum(mean, 1)))

#On définit la RMSE qui sert quantifier l'erreur: valeur moyenne de différence

def root_mean_square_error(y_pred, y_true):
    return math.sqrt(mean_squared_error(y_pred, y_true))

'ok'
```
